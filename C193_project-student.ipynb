```python
#Task 1 
#Import all the libraries and read cardiovascular.csv
import pandas as ps
import matplotlib.pyplot as plt

df = ps.read_csv("cardiovascular.csv")
df
```

# Average cholestrol found in various age groups

```python
#Task 2 
#Group by age and find the average cholesterol and make a dataframe out of it

group_df = df.groupby("age")["cholesterol"].mean().reset_index()
```

```python
#Task 3 
#Plot a line graph for various age group and their cholesterol 
fig = plt.subplots(figsize=(19, 8))

plt.plot(group_df["age"], group_df["cholesterol"], label="Average Cholesterol as per age", linewidth=3.0)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.xticks(rotation="vertical")

plt.title("Average Cholesterol as per age")

plt.legend()
plt.show()
```

Conclusion - The older your are the more cholestrol you will get.

# Correlation between systolic and diastolic blood pressure


```python
# Diastolic blood pressure - is the pressure in the arteries when the heart rests between beats
# Systolic blood pressure - the pressure in your arteries when your heart beats

#predefine code for image
from IPython.display import Image
Image(filename='blood pressure readings chart.jpg') 
#predefine code end
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe project out of it
group_df1 = df.groupby("age")["systolic_blood_pressure"].max().reset_index()
```

```python
#Task 5
#Group by age and find maximum diastolic blood pressure and create a dataframe project out of it
group_df2 = df.groupby("age")["diastolic_blood_pressure"].max().reset_index()
```

```python
#Task 6
#Plot a line graph to show a Correlation between systolic and diastolic blood pressure
group_df1 = df.groupby("age")["systolic_blood_pressure"].max().reset_index()
group_df2 = df.groupby("age")["diastolic_blood_pressure"].max().reset_index()
fig = plt.subplots(figsize=(19, 8))

plt.plot(group_df1["age"], group_df1["systolic_blood_pressure"], label="Maximum diastolic blood pressure as per age", linewidth=3.0)
plt.plot(group_df2["age"], group_df2["diastolic_blood_pressure"], label="Maximum Systolic blood pressure as per age", linewidth=3.0)

plt.xlabel("Age")
plt.ylabel("Systolic and diastolic blood pressure")
plt.xticks(rotation="vertical")

plt.title("Maximum Systolic and diastolic blood pressure as per age")

plt.legend()
plt.show()
```

Conclusion - both blood pressures vary form age to age but diatolic seems to happen more after the age of 40
