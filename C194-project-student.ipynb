# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
import pandas as ps
import matplotlib.pyplot as pyp
import numpy

df = ps.read_csv("framingham.csv")
df
```

```python
#Filter and make a new dataframe for non smokers
```

```python
#Group by age column and find average heart rate at different age
```

```python
#Filter and make a new dataframe for smokers
```

```python
#Group by age column and find average heart rate at different age
```

```python
#Plot a line graph to show the heart rate of smokers vs non smokers
smoker_df = df.loc[df["currentSmoker"] == 1]
smoker_df = smoker_df.groupby("age")["heartRate"].mean().reset_index()

non_smoker_df = df.loc[df["currentSmoker"] == 0]
non_smoker_df = non_smoker_df.groupby("age")["heartRate"].mean().reset_index()

pyp.figure(figsize=(16, 8))
pyp.title("Heart Rate of smokers and non smokers by age")

pyp.plot(smoker_df["age"], smoker_df["heartRate"], label="Smoker")
pyp.plot(non_smoker_df["age"], non_smoker_df["heartRate"], label="Non Smoker")

pyp.legend()
pyp.show()
```

Conslusion - Smokers have high heart rate

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv

#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
```

```python
#Group by age column and count the rows of TenYearCHD column
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years
CHD_df = df.loc[df["TenYearCHD"] == 1]
CHD_df = CHD_df.groupby("age")["TenYearCHD"].count().reset_index()

pyp.figure(figsize=(16, 8))
pyp.title("Number of people having a chance of coronary heart disease in 10 years per age group")

pyp.plot(CHD_df["age"], CHD_df["TenYearCHD"], label="People with CHD in 10 years")

pyp.legend()
pyp.show()
```

Conslusion - People around 60-65 have high likelihood of CHD in 10 years
