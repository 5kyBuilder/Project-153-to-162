```python
import pandas as ps
from matplotlib import pyplot as pyp

dataframe = ps.read_csv("bmi.csv")
dataframe = dataframe.dropna()
dataframe
```

```python
#Task 1
#How many people are underweight and create a dataframe out of it
DATA = dataframe.loc[dataframe['bmi'] <= 18.5]
groupbyname = DATA.groupby("gender")["bmi"].mean().reset_index()
fig = pyp.subplots(figsize=(16, 10))

pyp.xticks(rotation="vertical")
pyp.title("BMI Of underweight men and women")
pyp.xlabel("Gender")
pyp.ylabel("BMI")

pyp.bar(groupbyname["gender"], groupbyname["bmi"], width=0.4, color=["black", "brown", "pink", "magenta", "blue", "cyan", "red", "maroon", "purple", "orange", "yellow", "lime"])
```

```python
#Task 2
#How many people have normal weight and create a dataframe out of it
groupbyname = dataframe.groupby("gender")["bmi"].mean().reset_index()

fig = pyp.subplots(figsize=(16, 10))

pyp.xticks(rotation="vertical")
pyp.title("BMI Of average weight men and women")
pyp.xlabel("Gender")
pyp.ylabel("BMI")

pyp.bar(groupbyname["gender"], groupbyname["bmi"], width=0.4, color=["black", "brown", "pink", "magenta", "blue", "cyan", "red", "maroon", "purple", "orange", "yellow", "lime"])
```

```python
#Task 3
#Plot a pie chart as per the percentage of people who are underweight and healthy. ]
DATA = dataframe.loc[dataframe['bmi'] <= 18.5]
DATATWO = dataframe.loc[dataframe['bmi'] > 18.5]
DATATWO = dataframe.loc[dataframe['bmi'] < 25]

values = [len(DATA["bmi"]),len(DATATWO["bmi"])]

fig = pyp.subplots(figsize=(16, 10))

labels = ["Underweight", "Average Weight"]

pyp.title("BMI Of average weight and underweight")
pyp.xlabel("Gender")
pyp.ylabel("BMI")

pyp.pie(values,labels=labels, autopct='%0.2f%%', radius=1)
pyp.show()
```

```python
#Task 4
#Group by the gender from underweight dataframe and create another data frame out of it

groupbyname = DATA.groupby("gender")["gender"].count().reset_index(name="number")
print(groupbyname)
```

```python
#Task 5
#Plot a pie chart as per the percentage of male and female who are underweight

fig = pyp.subplots(figsize=(16, 10))

pyp.title("Percentage of underweight men and women")
pyp.xlabel("Gender")
pyp.ylabel("BMI")

pyp.pie(groupbyname["number"], labels=groupbyname["gender"], autopct='%0.2f%%', radius=1)
pyp.show()
```

```python
#Task 6
#Group by the gender from healthy weight dataframe and create another data frame out of it

groupbyname = DATATWO.groupby("gender")["gender"].count().reset_index(name="number")
print(groupbyname)
```

```python
#Task 7
#Plot a pie chart as per the percentage of male and female who are healthy
fig = pyp.subplots(figsize=(16, 10))

pyp.title("Percentage of average weight men and women")
pyp.xlabel("Gender")
pyp.ylabel("BMI")

pyp.pie(groupbyname["number"], labels=groupbyname["gender"], autopct='%0.2f%%', radius=1)
pyp.show()
```
