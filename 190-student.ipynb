```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as pyp

bmi_df = pd.read_csv('bmi.csv')

top_five = bmi_df.head(5)

age = top_five["Age"]
bmi = top_five["BMI"]

pyp.xlabel("Age")
pyp.ylabel("BMI")

pyp.bar(age, bmi, width=0.4, color=("blue", "red", "green", "black", "yellow"))
```
